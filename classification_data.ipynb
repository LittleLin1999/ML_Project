# Data preparation


```python
dataset = obj[1].data[(Teff_tot > 0) & (LOGG_tot>-8000) & (FEH_tot>-8000) & 
             (abs(bv_tot) < 2000) &  (abs(ug_tot) < 2000) &  
              (abs(gr_tot) < 2000) &  (abs(ri_tot) < 2000) & 
              (abs(iz_tot) < 2000)]
print(obj[1].data.shape)
print(dataset.shape)
```

```python
# split samples into training set & test set
import astroML
import numpy as np

features = ['Teff','logg','FeH','b-v','u-g','g-r','r-i','i-z']
X = np.empty((len(dataset),8),dtype = float)
X[:,0] = dataset['TEFF_ADOP']
X[:,1] = dataset['LOGG_ADOP']
X[:,2] = dataset['FEH_ADOP']
X[:,3] = dataset['BV']
X[:,4] = dataset['UG']
X[:,5] = dataset['GR']
X[:,6] = dataset['RI']
X[:,7] = dataset['IZ']

labels = list(set(dataset['SPECTYPE_SUBCLASS']))
y = np.zeros((len(dataset)),dtype=float)

for i in range(len(labels)):
    y[dataset['SPECTYPE_SUBCLASS']==labels[i]]=i
   
    
Group=np.full(len(labels),None)
for i  in range(len(labels)):
    Group[i] = X[y==i]
print([Group[i].shape for i in range(len(labels))])

plt.figure(figsize=(10,5))
plt.bar(range(len(labels)),[Group[i].shape[0] for i in range(len(labels))])
plt.title('Original Distribution')
plt.savefig('Original_Distribution.png',dpi=300)
```

### 

```python
Selection = []
new_labels = []
for sub in range(len(Group)):
    if Group[sub].shape[0] > 100:
        Selection.append(Group[sub])
        new_labels.append(labels[sub])
print(len(Selection))
print([(new_labels[i],len(Selection[i])) for i in range(len(Selection))])    

plt.figure(figsize=(20,5))
plt.bar(new_labels,[len(Selection[i]) for i in range(len(Selection))])
plt.title('new Distribution ')
plt.savefig('new_Distribution.png',dpi=300)
```

### train setlabelsample

```python
X_train = np.array(Selection[0][0].reshape(1,8))
X_test = np.array(Selection[0][0].reshape(1,8))
y_train = np.array([0])
y_test = np.array([0])    # labelsnew_labels
for j in range(len(Selection)):
    (subX_train,subX_test),(suby_train,suby_test) = split_samples(Selection[j], np.array([j]*len(Selection[j])), fractions=[0.75, 0.25], random_state=None)
    X_train = np.concatenate((X_train,subX_train),axis=0)
    X_test = np.concatenate((X_test,subX_test),axis=0)
    y_train = np.concatenate((y_train,suby_train),axis=0)  
    y_test = np.concatenate((y_test,suby_test),axis=0)                                                               
```

# Extra Tree Classifier

### 10 estimators

```python
# Training
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

forest = ExtraTreesClassifier(n_estimators=10, random_state=0)
forest.fit(X_train, y_train)

importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances for 10 estimators")
plt.bar(range(X.shape[1]), importances[indices],
       color="pink", yerr=std[indices], align="center")

ranged_features = [ features[f] for f in indices]
plt.xticks(range(X.shape[1]), ranged_features)
plt.xlim([-1, X.shape[1]])
plt.savefig('./classification/extraTree10_ranged_features.png',dpi=300)
plt.show()
```

```python
#Predict
y_pred = forest.predict(X_test)
```

```python
TF = (y_pred==y_test).astype(int)
Fail = len(TF[TF<0.5])/len(TF)
Succ = len(TF[TF>0.5])/len(TF)
print(Succ)
plt.bar(['Fail','Success'],[Fail,Succ],color='lightblue')
plt.title('Classification accuracy')
plt.savefig('ExtraTree10-ClassAccuracy.png',dpi=300)
```

```python
from sklearn.metrics import balanced_accuracy_score
score = balanced_accuracy_score(y_pred,y_test)
print(score)
```
